```python
import matplotlib.pyplot as plt
import numpy as np
```

<h2>No 1 : Multiple Subplots</h2>

Dengan data di bawah ini buatlah visualisasi seperti expected output :

```python
x = np.linspace(2*-np.pi, 2*np.pi, 200)
tan = np.tan(x)/10
cos = np.cos(x)
sin = np.sin(x)
```

```python
fig, ax = plt.subplots(nrows=1, ncols=3,squeeze=False,figsize=(17,6),sharey=True)
ax[0][0].plot(x,tan,label='tanx',color='b')
ax[0][1].plot(x,sin,label='sinx',color='r')
ax[0][2].plot(x,cos,label='cosx',color='g')
ax[0][0].set_title('Grafik tanx')
ax[0][1].set_title('Grafik sinx')
ax[0][2].set_title('Grafik cosx')

lines = []
labels = []

for ax in fig.axes:
    axLine, axLabel = ax.get_legend_handles_labels()
    lines.extend(axLine)
    labels.extend(axLabel)
    
fig.legend(lines, labels,           
          loc = 'upper center')
plt.show()
```

<h2>No 2 : Nested Axis</h2>

Dengan data di bawah ini, buatlah visualisasi seperti expected output :

```python
x = np.linspace(2*-np.pi, 2*np.pi, 100)
y = np.cos(x)
y2 = np.cos(x**2)
y3 = np.cos(x**3)
y4 = np.cos(x**4)
y5 = np.cos(x**5)
```

```python
fig, ax=plt.subplots(figsize=(13,8))
ax.plot(x,y,label='Grafik CosX',color='b')
ax.set_title('Grafik CosX')
kiri_atas=ax.inset_axes([-0.1,0.9,0.3,0.3])
kiri_atas.plot(x,y5,color='y')
kiri_atas.plot(x,y,color='b')
kiri_atas.set_title('Grafik CosX^5')
kiri_bawah=ax.inset_axes([-0.1,-0.1,0.3,0.3])
kiri_bawah.plot(x,y2,color='r')
kiri_bawah.plot(x,y,color='b')
kiri_bawah.set_title('Grafik CosX^2')
kanan_atas=ax.inset_axes([0.9,0.9,0.3,0.3])
kanan_atas.plot(x,y4, color='k')
kanan_atas.plot(x,y,color='b')
kanan_atas.set_title('Grafik CosX^4')
kanan_bawah=ax.inset_axes([0.9,-0.1,0.3,0.3])
kanan_bawah.plot(x,y3,color='g')
kanan_bawah.plot(x,y,color='b')
kanan_bawah.set_title('Grafik CosX^3')
plt.show()
```
